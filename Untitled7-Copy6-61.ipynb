```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
y=df['Longitude']
x=df1.drop(['Longitude','Abbreviation','Agricultural Land( %)','Land Area(Km2)','Armed Forces size','Capital/Major City','Co2-Emissions','CPI','Out of pocket health expenditure','Physicians per thousand','Population','Population: Labor force participation (%)','Tax revenue (%)','Total tax rate','Unemployment rate','Urban_population','CPI Change (%)','Currency-Code','Fertility Rate','Forested Area (%)','Gasoline Price','GDP','Gross primary education enrollment (%)','Gross tertiary education enrollment (%)','Infant mortality','Largest city','Life expectancy','Maternal mortality ratio','Minimum wage','Official language','Density\n(P/Km2)'],axis=1)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
print(x_train)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
model.intercept_
```

```python
coeff=pd.DataFrame(model.coef_,x.columns,columns=["Coefficient"])
coeff
```

```python
prediction=model.predict(x_test)
plt.scatter(y_test,prediction)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.linear_model import Ridge,Lasso
```

```python
rr = Ridge(alpha=10)
rr.fit(x_train,y_train)
```

```python
rr.score(x_test,y_test)
```

```python
la = Lasso(alpha=10)
la.fit(x_train,y_train)
```

```python
la.score(x_test,y_test)
```

```python
from sklearn.linear_model import ElasticNet
en=ElasticNet()
en.fit(x_train,y_train)
print(en.coef_)
print(en.intercept_)
print(en.predict(x_test))
print(en.score(x_test,y_test))
from sklearn import metrics
print("Mean Absolute Error:",metrics.mean_absolute_error(y_test,prediction))
print("Mean Squared Error:",metrics.mean_squared_error(y_test,prediction))
print("Root Mean Squared Error:",np.sqrt(metrics.mean_squared_error(y_test,prediction)))
```
